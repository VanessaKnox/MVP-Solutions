```python
AllAges_2009_2012_ZipCodes['Normalized'] = (AllAges_2009_2012_ZipCodes['Difference']+abs(AllAges_2009_2012_ZipCodes['Difference'].min()))/spread
AllAges_2009_2012_ZipCodes
```
